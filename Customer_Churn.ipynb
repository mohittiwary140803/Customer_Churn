```python
#Copy of the initial dataset for manipulation and preprocessing
df_copy = df.copy()
```

```python
#Converting the TotalCharges datatype to numeric from object
df_copy.TotalCharges = pd.to_numeric(df.TotalCharges, errors = 'coerce')
df_copy.isnull().sum()
```

```python
#As there are 11 missing values in the TotalCharges column, let's handle it.
df_copy[df_copy['TotalCharges'].isnull() == True]
```

```python
#Percentage of missing values in the TotalCharges column
len(df_copy[df_copy['TotalCharges'].isnull() == True])/len(df_copy['TotalCharges'])*100
```

```python
# As only 0.15% datas are missing so we can ignore them as it won't hamper our preprocessing and model selection
#Removing missing values
df_copy.dropna(inplace = True)
df_copy.isnull().sum()
```

```python
#Dividing customers based tenure into bins,
#e.g.- for tenure < 12 months , assign a bin 1-12
#for tenure 12 to 24 months, assign a bin 13-24 and so on
#Finding the maximum tenure
df_copy['tenure'].max()
```

```python
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]
df_copy['tenure_group'] = pd.cut(df_copy.tenure, range(1,80,12), right = False, labels = labels)
```

```python
df_copy['tenure_group'].value_counts()
```

```python
#Dropping unwanted columns - customerID and tenure
df_copy.drop(['customerID','tenure'], axis=1, inplace=True)
```

```python
df_copy.head()
```

```python
#PLOT DISTRIBUTION OF INDIVIDUAL PREDICTIONS BY CHURN
for i, predictor in enumerate(df_copy.drop(['Churn','TotalCharges','MonthlyCharges'], axis=1)):
  plt.figure(i)
  sns.countplot(data = df_copy, x = predictor, hue='Churn')
```

```python
#Coverting the binary categorical variable into binary numeric variable i.e., Yes - 1 & No - 0
df_copy['Churn'] = df_copy['Churn'].map({'Yes' : 1, 'No' : 0})
```

```python
df_copy.head()
```

```python
#Converting all the categorical variables intp dummy variables.
df_copy_dummies = pd.get_dummies(df_copy)
```

```python
df_copy_dummies.head()
```

```python
#Relationship between MonthlyCharges and TotalCharges
sns.lmplot(data=df_copy_dummies, x= 'MonthlyCharges', y= 'TotalCharges', fit_reg=False)
#We can see as the MonthlyCharges are increasing , TotalCharges are also increasing.
```

```python
#CHURN BY MonthlyCharges
monthly = sns.kdeplot(df_copy_dummies.MonthlyCharges[(df_copy_dummies['Churn']==0)], color='Red', fill=True)
monthly = sns.kdeplot(df_copy_dummies.MonthlyCharges[(df_copy_dummies['Churn']==1)], color='Blue',fill=True)
monthly.legend(['No Churn','Churn'],loc='upper right')
monthly.set_title('Monthly Charges by CHURN')
```

```python
#TotalCharges by CHURN
total = sns.kdeplot(df_copy_dummies.TotalCharges[(df_copy_dummies['Churn']==1)], color='Blue', fill = True)
total = sns.kdeplot(df_copy_dummies.TotalCharges[(df_copy_dummies['Churn']==0)], color='Red', fill = True)
total.set_title('Total Charges by CHURN')
total.legend(['Churn','No Churn'],loc = 'upper right')
```

## **MODEL** **BUILDING**

```python
df_copy_dummies
```

```python
X = df_copy_dummies.drop('Churn', axis = 1)
X
```

```python
y = df_copy_dummies['Churn']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

# **DECISION TREE CLASSIFIER**

```python
#DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier
dec_tree = DecisionTreeClassifier(criterion='gini', random_state=0, max_depth = 6, min_samples_leaf=8)
dec_tree.fit(X_train, y_train)
```

```python
y_pred_dt = dec_tree.predict(X_test)
print(y_pred_dt)
```

```python
from sklearn.metrics import recall_score, confusion_matrix, classification_report
print(dec_tree.score(X_test, y_test)*100)
print(confusion_matrix(y_test, y_pred_dt))
```

```python
print(classification_report(y_test, y_pred_dt, labels=[0,1]))
```

```python
#AS THE ACCURACY IS COMING OUT TO BE VERY LESS BCOZ OF IMBALANCED DATASET, SO WE ARE GOING TO IMPLEMENT SMOTEENN i.e., UpSampling + Edited Nearest Neighbors
```

Implementing SMOTEENN (UPSAMPLING + ENN) in Decision Tree Classifier

```python
from imblearn.combine import SMOTEENN
sm_dt = SMOTEENN()
X_sampled_dt , y_sampled_dt = sm_dt.fit_resample(X,y)
```

```python
Xr_train_dt , Xr_test_dt, yr_train_dt, yr_test_dt = train_test_split(X_sampled_dt, y_sampled_dt, test_size=0.2)
```

```python
smote_dt = DecisionTreeClassifier(criterion='gini',random_state=100, max_depth=6, min_samples_leaf=8)
smote_dt.fit(Xr_train_dt, yr_train_dt)
```

```python
yr_pred_dt = smote_dt.predict(Xr_test_dt)
yr_pred_dt
```

```python
print(smote_dt.score(Xr_test_dt, yr_test_dt)*100)
print(confusion_matrix(yr_test_dt, yr_pred_dt))
```

```python
print(classification_report(yr_test_dt, yr_pred_dt, labels=[0,1]))
```

# **K-NEAREST NEIGHBORS CLASSIFIER**

```python
#KNN CLASSIFIER
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p=2)
knn.fit(X_test, y_test)
```

```python
y_pred_knn= knn.predict(X_test)
print(y_pred_knn)
```

```python
from sklearn.metrics import recall_score, confusion_matrix, classification_report
print(knn.score(X_test, y_test)*100)
print(confusion_matrix(y_test, y_pred_knn))
```

```python
print(classification_report(y_test, y_pred_knn, labels=[0,1]))
```

```python
from imblearn.combine import SMOTEENN
sm_knn = SMOTEENN()
X_sampled_knn , y_sampled_knn = sm_knn.fit_resample(X,y)
```

```python
yr_pred_rfc = smote_rfc.predict(Xr_test_rfc)
yr_pred_rfc
```

```python
from sklearn.metrics import recall_score, confusion_matrix, classification_report
print(smote_rfc.score(Xr_test_rfc, yr_test_rfc)*100)
print(confusion_matrix(yr_test_rfc, yr_pred_rfc))
```

```python
print(classification_report(yr_test_rfc, yr_pred_rfc, labels=[0,1]))
```
